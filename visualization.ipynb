# Embeddings visualization

```python
import importlib
import embedding_models
importlib.reload(embedding_models)
from embedding_models import EmbeddingModel
import tqdm as notebook_tqdm

texts = [
    "Artificial intelligence is revolutionizing finance",
    "Stock markets drop after FED announcement",
    "Tesla unveils new battery with record range",
    "Apple launches iPhone 16 with new features",
    "The job market in France is seeing an increase in hiring",
    "AI and cybersecurity: A new era of data protection",
    "Artificial intelligence is revolutionizing economy",
]
```

```python
# Initialize an SBERT model
model = EmbeddingModel(method="sbert")
model.fit(texts)
```

```python
# PCA Visualization
model.visualize_embeddings(texts, method="pca", labels=[i for i in range(len(texts))])
```

```python
# t-SNE Visualization
model.visualize_embeddings(texts, method="tsne", labels=[i for i in range(len(texts))])
```

```python
# t-SNE scaled Visualization
model.visualize_embeddings(texts, method="tsne scaled", labels=[i for i in range(len(texts))])
```

# 📊 **Analysis of Embedding Visualizations (PCA & t-SNE)**

---

## 📌 **1️⃣ PCA Results**

✅ PCA provides a **global structure**, where similar texts tend to be closer.  
✅ The points **retain some distance relationships**, meaning texts that are semantically different are well-separated.  
✅ Clusters are **less clear** compared to t-SNE, as PCA prioritizes variance explanation rather than neighborhood preservation.  

### **Notable Clusters**
- The Tesla & Apple thing are close, which makes sense as both refer to **technology product announcements**.
- **"Artificial intelligence" & "AI and cybersecurity"** are close together as they mention AI.
- **"Stock markets drop"** is isolated, indicating that it is semantically distinct from other topics.

---

## 📌 **2️⃣ t-SNE Results (Perplexity = 3)**

✅ Unlike PCA, t-SNE **prioritizes local structure**, leading to a more **clustered representation**.  
✅ The best perplexity value was found to be **3**, balancing global and local relationships.  
✅ Some **outliers** are still present, meaning that some embeddings might not be optimally placed.

### **Notable Clusters**
- **"Artificial intelligence" & "AI and cybersecurity"** remain close, suggesting embeddings capture the semantic meaning.
- **"Tesla unveils" & "Apple launches"** are still paired but slightly further apart compared to PCA.
- **"The job market"** is positioned differently than in PCA, maybe it indicates t-SNE interprets its relationships differently.
- The **Finance** & **Economy** lines, are quite far from each other. It leads me to conclude that PCA is better for visualization in this case. 

📌 **t-SNE is better for spotting clusters, but global distances may not be meaningful.**  

---

## 📌 **3️⃣ t-SNE Scaled Results (Perplexity = 2)**

### **Key Observations**
✅ Scaling embeddings before t-SNE **improves stability** and **prevents distortion** of distances.  
✅ I don't really feel like distance are **more balanced** compared to the non-scaled TSNE.
